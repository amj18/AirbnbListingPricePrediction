## Import data from URL and unzip to local directory

```python
curr_dir = os.getcwd()

if not os.path.exists('datasets'):
    os.mkdir('datasets')

url = "http://data.insideairbnb.com/united-kingdom/england/london/2019-07-10/data/listings.csv.gz"
# urllib.request.urlretrieve(url, os.path.join(curr_dir, "datasets", "listings.csv.gz"))
```

```python
with gzip.open(curr_dir+r"/datasets/listings.csv.gz", "rb") as f_in:
    with open(curr_dir+r"/datasets/listings_raw.csv", "wb") as f_out:
        shutil.copyfileobj(f_in, f_out)
```

```python
data_path = curr_dir+r"/datasets/listings_raw.csv"
df = pd.read_csv(data_path)
```
